# 演習３　量子化誤差の見積もり

## 準備
適当な信号（例えば周期64の正弦波を）生成し，この信号を量子化ビット数$r$で量子化したときの量子化誤差の${\rm SNR}$と$r$の間の傾きが約$6$[dB/bit]になることを確認せよ．ここで，$r$は例えば1ビットから11ビットまで変動させるなどするとよい．

## ヒント
* 常用対数： `np.log10`  
* 分散： `np.var`  
* フロア関数： `np.floor`  
* 1から11までの整数列をつくる: `range(1:12)`
* 空のリストを作る：`snr_list=[]`  
* リストに値(valとする）を追加： `snr_list.append(val)`  

## 進め方
以下のようなステップで進めるといいでしょう。  
1. 固定した量子化ビット長（$r=2$ など）で信号`x`を量子化する
1. 固定した量子化ビット長($r=2$ など）で分散$\sigma^2_e$を計算する
1. 上の2.と3.を（$r$についての）`for`文のループにいれて結果(SNR)をリストに保存
1. 横軸$r$、縦軸SNRのグラフを書く
1. グラフの傾きが概ね6になることが確認する


```
import numpy as np 
import matplotlib.pyplot as plt 
t = np.linspace(0, 255, 256)
x = np.sin(t*2*np.pi/64)

plt.plot(t,x)
plt.show()

```
