```
def D(im):
    h, w = im.shape 
    dh = np.concatenate([im[1:, :] - im[0:-1, :], \
        np.zeros([1, w])], axis=0) 
    dw = np.concatenate([im[:, 1:] - im[:, 0:-1], \
        np.zeros([h, 1])], axis=1) 
    return np.dstack([dh, dw])

#Dtの定義
```
